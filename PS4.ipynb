#Ouestion 4)Write a program in C# Sharp to separate odd and even integers in separate arrays. Go to the editor
Test Data :
Input the number of elements to be stored in the array :5
Input 5 elements in the array : element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56 element - 4 : 32 Expected Output:
The Even elements are: 42 56 32
The Odd elements are : 25 47

```python
using System;

namespace FarukFurkan.KOC.BDAT1004PS4  #Virtual Studio
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.WriteLine("Test Data");



            Console.WriteLine("Input the number of elements to be stored in the array : ");
            # arrayLength is equal to user input
            int arraylength = Convert.ToInt32(Console.ReadLine());
            Console.WriteLine("Input {0} elements in the array: ", arraylength);
            # Array that consist of the user inputs
            int[] array = new int[arraylength];
            #Assigns the user inputs to the array's index
            for (var i = 0; i < arraylength; i++)
            {



                int a = Convert.ToInt32(Console.ReadLine());
                Console.WriteLine("element - " + i + " : " + a);
                array[i] = a;
            }
            # Arrays for even and odd numbers
            int[] evenArray = new int[arraylength];
            int[] oddArray = new int[arraylength];
            # Count variable to declare the index of the arrays
            int count1 = 0;
            int count2 = 0;



            # Checks the even and odd index of the 'array'
            for (int i = 0; i < array.Length; i++)
            {
                if (array[i] % 2 == 0)
                {

                    evenArray[count1] = array[i];
                    count1++;



                }
                else
                {
                    oddArray[count2] = array[i];
                    count2++;
                }



            }



            #Prints the even and odd index
            Console.WriteLine("The Even elements are: ");
            for (var i = 0; i < evenArray.Length; i++)
            {

                if (evenArray[i] != 0)
                {
                    Console.Write(evenArray[i] + " ");
                }
            }
            Console.WriteLine(" ");
            Console.WriteLine("The Odd elements are :");
            for (var i = 0; i < oddArray.Length; i++)
            {

                if (oddArray[i] != 0)
                {
                    Console.Write(oddArray[i] + " ");
                }
            }




        }
    }
}
```
